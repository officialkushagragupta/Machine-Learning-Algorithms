```python
table_row = []
```

```python
for row in row_data:
    curr_row = []
    r_data = row.find_all('td', {})
    for data in r_data:
        curr_row.append(data.text)
    table_row.append(curr_row)
```

```python
file_name = 'countries.csv'
```

```python
col_tittle = []
for col in first_row:
    col_tittle.append(col.text)
```

```python
headings = first_row.find_all('th')
```

```python
headings
```

```python
col_tittle = []
for col in headings:
    col_tittle.append(col.text)
```

```python
col_tittle
```

```python
with open(file_name, 'w', encoding='utf-8') as file:
    head_str = ','.join(col_tittle) + '\n'
    file.write(head_str)
    
    for row in table_row:
        row_str = ""
        row_str = ','.join(row)
        file.write(row_str)
```

```python
file_name = 'countries_separated_comma.csv'
```

```python
with open(file_name, 'w', encoding = 'utf-8') as f:
    head_str = ','.join(col_tittle) + '\n'
    print(head_str)
    f.write(head_str)
    
    for row in table_row[:-1]:
        row_str = ""
        
        for w in row:
            w = w.replace(',', '')
            row_str += w + ','
        row_str = row_str[:-1]
        row_str += '\n'
        f.write(row_str)
```
