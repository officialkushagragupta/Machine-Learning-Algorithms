```python
plt.plot(X, y1, color = 'red', label = 'Oil', linestyle = '--')
plt.plot(X, y2, color = 'green', label = 'Gold')
plt.xlabel('Time')
plt.ylabel('Prices')
plt.legend(title = 'Prices of Comodities')
plt.title('Time vs Prices')
```

```python
plt.scatter(X, y1, color = 'red', label = 'Oil', marker= '^')
plt.scatter(X, y2, color = 'green', label = 'Gold', marker = '+')
plt.xlabel('Time')
plt.ylabel('Prices')
plt.legend(title = 'Prices of Comodities')
plt.title('Time vs Prices')
```

```python
plt.bar(['2005', '2010', '2015', '2020', '2023'], [2, 5, 7, 7, 10])
plt.xlabel('Time')
plt.ylabel('Price')
```

```python
X = np.array([0, 1, 2])
plt.bar(X, [10, 15, 16], width = 0.4, color = 'green', label = 'Gold', tick_label = ['2005', '2010', '2015'])
plt.bar(X+0.4, [11, 15, 20], width = 0.4, color = 'red', label = 'Oil', tick_label = ['2005', '2010', '2015'])
plt.bar(X + 0.2, [0, 0, 0], width=0.4, color = 'blue', label = 'nil', tick_label = ['2005', '2010', '2015'])
```
